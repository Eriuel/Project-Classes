```python
#Jonathan Colbourne
#ID: 000897327

from tabulate import tabulate

class doctor():
  def formatDrInfo(self):
    pass # I pull the data from the txt, formatting is an uneeded step.

  def enterDrInfo(self):
    doctor = []
    doctor.append(input("ID: \n"))
    doctor.append(input("Name: \n"))
    doctor.append(input("Specialization: \n"))
    doctor.append(input("Working Time: \n"))
    doctor.append(input("Qualification: \n"))
    doctor.append(input("Room Number: \n"))
    return '_'.join(doctor)

  def readDoctorsFile(self):
    txt = open('/content/doctors.txt', 'r').read()
    dataFormated = txt.split('\n')
    return dataFormated

  def searchDoctorById(self):
    id = input('ID: \n')
    data = self.readDoctorsFile()
    ls = []
    ls.append(data[0].split('_'))
    e = 'Info not found.'
    for x in data:
      if (x.find(id, 0, 3) != -1):
        e = x
    ls.append(e.split('_'))
    print(tabulate(ls, headers='firstrow'))

  def searchDoctorByName(self):
    name = input('Name: \n')
    data = self.readDoctorsFile()
    ls = []
    ls.append(data[0].split('_'))
    e = 'Info not found.'
    for x in data:
      if (x.find(name) != -1):
        e = x
    ls.append(e.split('_'))
    print(tabulate(ls, headers='firstrow'))

  def displayDoctorInfo(self):
    data = self.readDoctorsFile()
    formatedData = []
    for x in data:
      formatedData.append(x.split('_'))
    return formatedData

  def editDoctorInfo(self):
    id = input('ID: \n')
    data = self.readDoctorsFile()
    editor = open('/content/doctors.txt', 'w')
    newData = []
    for x in data:
      if (x.find(id, 0, 4) != -1):
        newData.append(self.enterDrInfo())
      else:
        newData.append(x)
    newDataJ = "\n".join(newData)
    editor.writelines(newDataJ)


  def displayDoctorsList(self):
    print(tabulate(self.displayDoctorInfo(), headers='firstrow'))

  def addDrToFile(self):
    data = self.readDoctorsFile()
    newData = '\n' + self.enterDrInfo()
    editor = open('/content/doctors.txt', 'w')
    dataJ = "\n".join(data)
    editor.writelines(dataJ)
    editor.writelines(newData)

class facility():
  def addFacility(self):
    txt = open('/content/facilities.txt', 'a')
    uI = '\n' + input("Facility name: \n")
    txt.write(uI)

  def displayFacilities(self):
    data = open('/content/facilities.txt', 'r')
    print(data.read())

class laboratory():
  def addLabToFile(self):
    data = self.readLaboratoriesFile()
    newData = '\n' + self.enterLaboratoryInfo()
    editor = open('/content/laboratories.txt', 'w')
    dataJ = "\n".join(data)
    editor.writelines(dataJ)
    editor.writelines(newData)

  def displayLabsList(self):
    data = self.readLaboratoriesFile()
    formatedData = []
    for x in data:
      formatedData.append(x.split('_'))
    print(tabulate(formatedData, headers='firstrow'))

  def formatDrInfo(self):
    pass

  def enterLaboratoryInfo(self):
    facility = []
    facility.append(input('Lab Name: \n'))
    facility.append(input('Cost: \n'))
    return '_'.join(facility)

  def readLaboratoriesFile(self):
    txt = open('/content/laboratories.txt', 'r').read()
    dataFormated = txt.split('\n')
    return dataFormated

# I was going to use inheritance, but there are too many name differences and etc.
class patient():
  def enterPatientInfo(self):
    patient = []
    patient.append(input('Pid: \n'))
    patient.append(input('Name: \n'))
    patient.append(input('Disease: \n'))
    patient.append(input('Gender\n'))
    patient.append(input('Age: \n'))
    return '_'.join(patient)

  def readPatientInfo(self):
    txt = open('/content/patients.txt', 'r').read()
    dataFormated = txt.split('\n')
    return dataFormated

  def searchPatientById(self):
    id = input('ID: \n')
    data = self.readPatientInfo()
    ls = []
    ls.append(data[0].split('_'))
    e = 'Info not found.'
    for x in data:
      if (x.find(id, 0, 3) != -1):
        e = x
    ls.append(e.split('_'))
    print(tabulate(ls, headers='firstrow'))

  def displayPatientInfo(self):
    data = self.readPatientInfo()
    formatedData = []
    for x in data:
      formatedData.append(x.split('_'))
    return formatedData

  def editPatientInfo(self):
    id = input('ID: \n')
    data = self.readPatientInfo()
    editor = open('/content/patients.txt', 'w')
    newData = []
    for x in data:
      if (x.find(id, 0, 4) != -1):
        newData.append(self.enterPatientInfo())
      else:
        newData.append(x)
    newDataJ = "\n".join(newData)
    editor.writelines(newDataJ)

  def displayPatientList(self):
    print(tabulate(self.displayPatientInfo(), headers='firstrow'))

  def addPatientToFile(self):
    data = self.readPatientInfo()
    newData = '\n' + self.enterPatientInfo()
    editor = open('/content/patients.txt', 'w')
    dataJ = "\n".join(data)
    editor.writelines(dataJ)
    editor.writelines(newData)

class management():
  def DisplayMenu(self):
    while True:
      print('Welcome to Alberta Hospital (AH) Managment system')
      print('Select from the following options, or select 0 to stop:')
      print('1 - 	Doctors')
      print('2 - 	Facilities')
      print('3 - 	Laboratories')
      print('4 - 	Patients ')
      userIn = input()
      if (userIn == '1'):
        while True:
          print('Doctors Menu:')
          print('1 - Display Doctors list')
          print('2 - Search for doctor by ID')
          print('3 - Search for doctor by name')
          print('4 - Add doctor')
          print('5 - Edit doctor info')
          print('6 - Back to the Main Menu')
          uI = input()
          if (uI == '1'):
            doctors.displayDoctorsList()
          if (uI == '2'):
            doctors.searchDoctorById()
          if (uI == '3'):
            doctors.searchDoctorByName()
          if (uI == '4'):
            doctors.addDrToFile()
          if (uI == '5'):
            doctors.editDoctorInfo()
          if (uI == '6'):
            break
      if (userIn == '2'):
        while True:
          print('Facilities Menu:')
          print('1 - Display Facilities list')
          print('2 - Add Facility')
          print('3 - Back to the Main Menu')
          uI = input()
          if (uI == '1'):
            buildings.displayFacilities()
          if (uI == '2'):
            buildings.addFacility()
          if (uI == '3'):
            break
      if (userIn == '3'):
        while True:
          print('Laboratories Menu:')
          print('1 - Display laboratories list')
          print('2 - Add laboratory')
          print('3 - Back to the Main Menu')
          uI = input()
          if (uI == '1'):
            lab.displayLabsList()
          if (uI == '2'):
            lab.addLabToFile()
          if (uI == '3'):
            break
      if (userIn == '4'):
        while True:
          print('Patients Menu:')
          print('1 - Display patients list')
          print('2 - Search for patient by ID')
          print('3 - Add patient')
          print('4 - Edit patient info')
          print('5 - Back to the Main Menu')
          uI = input()
          if (uI == '1'):
            patients.displayPatientList()
          if (uI == '2'):
            patients.searchPatientById()
          if (uI == '3'):
            patients.addPatientToFile()
          if (uI == '4'):
            patients.editPatientInfo()
          if (uI == '5'):
            break
      if (userIn == '0'):
        break

doctors = doctor()
buildings = facility()
lab = laboratory()
patients = patient()
manage = management()

manage.DisplayMenu()
```
